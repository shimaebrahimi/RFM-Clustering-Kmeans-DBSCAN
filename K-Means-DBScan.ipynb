```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.cluster import KMeans
import sklearn
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
data=pd.read_excel("Online_Retail.xlsx")
```

```python
data
```

```python
data.info()
```

```python
data=data[data['CustomerID'].notnull()]
```

```python
data.dropna()
```

```python
data
```

```python
data['InvoiceDay']=data['InvoiceDate'].apply(lambda x:dt.datetime(x.year,x.month,x.day))
```

```python
data
```

```python
pin_data=max(data['InvoiceDay'])+dt.timedelta(1)
```

```python
pin_data
```

```python
data['total_sum']=data['Quantity']*data['UnitPrice']
```

```python
data
```

```python
rfm=data.groupby('CustomerID').agg({
    'InvoiceDay':lambda x:(pin_data-x.max()).days,
    'InvoiceNo':'count',
    'total_sum':'sum'

})
```

```python
rfm
```

```python
rfm.rename(columns={
    'InvoiceDay':'Recency',
    'InvoiceNo':'Frequency',
    'total_sum' : 'Monetary'
},inplace=True)
```

```python
rfm
```

```python
r_labels=range(4,0,-1)
r_group=pd.qcut(rfm['Recency'],q=4,labels=r_labels)
f_labels=range(1,5)
f_group=pd.qcut(rfm['Frequency'],q=4,labels=f_labels)
m_labels=range(1,5)
m_group=pd.qcut(rfm['Monetary'],q=4,labels=m_labels)
```

```python
rfm['R']=r_group.values
rfm['F']=f_group.values
rfm['M']=m_group.values
```

```python
rfm
```

```python
x=rfm[['R','F','M']]
```

```python
x
```

```python
dbs=DBSCAN(eps=0.3,min_samples=10)
```

```python
dbs.fit(x)
```

```python
dbs.labels_
```

```python
np.unique(dbs.labels_)
```

```python
sl=metrics.silhouette_score(x,labels=dbs.labels_)
```
